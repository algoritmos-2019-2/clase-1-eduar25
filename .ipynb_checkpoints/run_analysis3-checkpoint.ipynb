Next, import all the necesssary modules and features.

```python
# Import modules
import csv
import matplotlib.pyplot as plt
import scipy as sc
from scipy.fftpack import fft
from scipy.signal import butter, lfilter, filtfilt
from scipy.integrate import trapz, cumtrapz
from __future__ import division, print_function
import numpy as np
import wave
import sys
sys.path.insert(1, r'./../functions')  # add to pythonpath
```

Data was obtained using the Backyard Brains EMG Spiker Box. The recordings are saved as audio files in .wav format. So, the first thing we have to do is open the .wave files and extract the data.

```python
# Open wav file (electromyogram data)
record = wave.open('/Users/emckiernan/UNAM/cursos/medicionAnalisis/Registros/Diego/170316DE1ec/Dec0.wav', 'r') # load the data
```

```python
# Get wave data
frames = record.readframes(-1)
waveData = np.fromstring(frames, 'Int16')
```

From the recordings we can extract the number of recording channels, sampling rate, etc.

```python
# Get the number of channels, sample rate, etc.
numChannels = record.getnchannels()
sampleRate = record.getframerate()
sampleWidth = record.getsampwidth()
numFrames = record.getnframes()

print('There is %d channel.' % (numChannels))
print('The sampling rate is %d Hz.' % (sampleRate))
```

Now, calculate the time window from the raw data and the sampling rate.

```python
# Get time window
timeEMG=np.linspace(0, len(waveData)/sampleRate, num=len(waveData))
```

Now, let's plot the raw EMG signal, before any kind of processing.

```python
# Plot EMG signal
plt.figure(1)
plt.title('Electromiograma')
plt.xlabel(r'tiempo (s)')
plt.ylabel(r'voltaje ($\mu$V)')
plt.plot(timeEMG,waveData)
plt.xlim(0,max(timeEMG))
plt.show()
```

Now, let's process the raw data by applying a band-pass filter at 10-400 Hz.

```python
# band-pass Butterworth filter at 10-400Hz
freq = 1/np.mean(np.diff(timeEMG))
b, a = butter(2, ([10, 400]/(freq/2)), btype = 'bandpass')
dataf = filtfilt(b, a, waveData)
```

Plotting a small time window from the raw and filtered recordings allows us to compare and see the effectiveness of the band-pass filtering.

```python
# plot data
fig, (ax1,ax2) = plt.subplots(2, 1, sharex = True, sharey = False, figsize = (8, 7))
ax1.plot(timeEMG, waveData, 'r')
ax1.set_title('Raw data', fontsize=12)
ax1.set_xlim(30,35)
ax1.set_xlabel(r'tiempo (s)')
ax1.grid()
ax2.plot(timeEMG, dataf, 'b')
ax2.set_title('Band-pass filter at [10 400] Hz', fontsize=12)
ax2.set_xlim(30,35)
ax2.grid()
fig.text(0, 0.5, 'voltaje ($\mu$V)', ha='center', va='center',
         rotation='vertical', fontsize=12)
plt.locator_params(axis = 'both', nbins = 4)
plt.tight_layout(h_pad=.1)


plt.ylabel(r'voltaje ')
```

```python
# Rectify EMG signal
absSignal = np.absolute(waveData)
timeAbs=np.linspace(0, len(absSignal)/sampleRate, num=len(absSignal))

# Normalize by maximum amplitude
normSignal = absSignal/absSignal.max()


# Calculate the cumulative sum
sumSignal = np.cumsum(normSignal)

print(absSignal)
print(sumSignal[-1]/timeEMG[-1])
```

```python
#normalized power spectral density
ps = np.abs(np.fft.fft(normSignal))**2

time_step = 1. / sampleRate
freqs = np.fft.fftfreq(waveData.size, time_step)
idx = np.argsort(freqs)
goodinds = sc.where(freqs[idx]>0)[0]
F = freqs[idx][goodinds]
P = ps[idx][goodinds]

plt.plot(F, P/P.max())
plt.ylim(0,1e-2)
plt.xlim(0,1e3)
plt.xlabel('Frequency (Hz)')
```

```python
def analyzeEMG(file='/Users/emckiernan/UNAM/cursos/medicionAnalisis/Registros/Diego/170316DE1ec/Dec0.wav'):
    # open and extract data from wav file
    record = wave.open(file,'r')
    frames = record.readframes(-1)
    waveData = np.fromstring(frames, 'Int16')
    # calculate number of channels, sample rate, etc.
    numChannels = record.getnchannels()
    sampleRate = record.getframerate()
    sampleWidth = record.getsampwidth()
    numFrames = record.getnframes()
    print('There is %d channel.' % (numChannels))
    print('The sampling rate is %d Hz.' % (sampleRate))
    # calculate time window
    timeEMG=np.linspace(0, len(waveData)/sampleRate, num=len(waveData))
    # calculate frequency
    freq = 1/np.mean(np.diff(timeEMG))
    # band-pass Butterworth filter at 10-400Hz
    b, a = butter(2, ([10, 400]/(freq/2)), btype = 'bandpass')
    dataf = filtfilt(b, a, waveData)
    # rectify the filtered EMG signal
    absSignal = np.absolute(dataf)
    timeAbs=np.linspace(0, len(absSignal)/sampleRate, num=len(absSignal))
    # normalize the EMG signal
    normSignal = absSignal/absSignal.max()
    # calculate the cumulative sum over the signal
    sumSignal = np.cumsum(absSignal)
    print(absSignal)
    print(sumSignal[-1]/timeEMG[-1])
    # calculate normalized power spectral density
    ps = np.abs(np.fft.fft(absSignal))**2
    time_step = 1 / sampleRate
    freqs = np.fft.fftfreq(absSignal.size, time_step)
    idx = np.argsort(freqs)
    goodinds = sc.where(freqs[idx]>0)[0]
    F = freqs[idx][goodinds]
    P = ps[idx][goodinds]


    # Plot EMG signal
    plt.figure(1)
    plt.title('Electromiograma')
    plt.xlabel(r'tiempo (s)')
    plt.ylabel(r'voltaje ($\mu$V)')
    plt.plot(timeAbs,absSignal)
    
    # Plot power spectral density
    plt.figure(2)
    plt.title('Power spectral density')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel(r'V^2/Hz')
    plt.plot(F, P/P.max())
    plt.ylim(0,1e-2)
    plt.xlim(0,1e3)
 
    plt.show()
    
    return 
```
